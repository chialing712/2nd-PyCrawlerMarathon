### ① 取出今天所有的發文

```python
soup
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html5lib')
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find_all('a')[-1].text)
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
cap = []
news = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find(class_="date").text,d.find_all('a')[-1].text)
    cap.append(d.find_all('em')[-1].text)
    news.append(d.find_all('a')[-1].text)
```

```python
import pandas as pd
cat = pd.DataFrame({'article': news}, index = cap)
cat
```

### ③ 哪一個類別的文章最多

```python
#elimiate duplicated values
category = list(set(cap))
for i in category:
    print(i, cap.count(i))
```
